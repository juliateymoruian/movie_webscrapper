**6. What did you discover/what were your conclusions (i.e were your original assumptions confirmed?** 

My original conclusion was confirmed that audiences and critics have different priorities regarding what could be considered a "top-rated" film. My top findings include the following:  

1) While both critics and audiences prefer Thrillers and Dramas, critics prefer films with Romantic themes, whereas audiences lean more towards Fantasy and Comedy. 

2) Highest-grossing films do not correlate with highest-rated films. Franchise films (The Avengers, Lord of the Rings, Star Wars) were the highest grossing in the audience's top-rated films, but still received average audience ratings. 3) "Classic films" continue to be the highest-rated amongst both categories. Audiences tend to rank franchises higher (Lord of the Rings, Avengers, Star Wars). Films prior to 1975 remain in the critics top rankings. 

4) Rated R films typically get higher ratings from audiences and lower ratings from critics. On average critics rank films rated PG and G higher than audiences. 

5) Robert Downey Jr. is the common lead actor across the highest top grossing films whereas Tom Hanks is the lead actor in the most top grossing films. Elijah Wood is the common actor that has been in the highest rated films amongst audiences and Robert Dinero is the lead actor that is in the highest rated films amongst critics.
